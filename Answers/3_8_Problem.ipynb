**Problem 3.8** Consider replacing the ReLU activation function with (i) the Heaviside step function $\text{heaviside}[z]$, (ii) the hyperbolic tangent function $\text{tanh}[z]$, and (iii) the rectangular function $\text{rect}[z]$, where:

$$
\text{heaviside}[z] = \begin{cases}
0 & z < 0 \\
1 & z \ge 0
\end{cases}
$$

$$
\text{rect}[z] = \begin{cases}
0 & z < 0 \\
1 & 0 \le z \le 1 \\
0 & z > 1
\end{cases}
$$

Redraw a version of figure 3.3 for each of these functions. The original parameters were: $\bold{\phi} = \{ \phi_0, \phi_1, \phi_2, \phi_3, \theta_{10}, \theta_{11}, \theta_{20}, \theta_{21}, \theta_{30}, \theta_{31} \} = \{ -0.23, -1.3, 1.3, 0.66, -0.2, 0.4, -0.9, 0.9, 1.1, -0.7 \}$. Provide an informal description of the family of functions that can be created by neural networks with one input, three hidden units, and one output for each activation function.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the shallow neural network model
def relu(x):
  return np.maximum(0, x)

def heaviside(x):
  return 1 * (x > 0)

def tanh(x):
  return np.tanh(x)

def rect(x):
  return 1 * (x > 0) * (x < 1)

def f(x, activation):
  h1 = activation(-0.2 + 0.4*x)
  h2 = activation(-0.9 + 0.9*x)
  h3 = activation( 1.1 - 0.7*x)
  return -0.23 - 1.3*h1 + 1.3*h2 + 0.66*h3
```

```python
# Function to help plot the data
def plot(activation):
    fig,ax = plt.subplots()
    plt.xlim([-4,4])
    plt.ylim([-1,1])
    ax.set_xlabel('Input, $x$')
    ax.set_ylabel('Output, $y$')
    # Draw line
    x_line = np.arange(-4, 4,0.01)
    y_line = f(x_line, activation)
    plt.plot(x_line, y_line,'b-',lw=2)

    plt.show()
```

```python
plot(activation=relu)
```

```python
plot(activation=heaviside)
```

```python
plot(activation=tanh)
```

```python
plot(activation=rect)
```
