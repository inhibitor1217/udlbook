Given a linear regression with three hidden units at fixed joints at 0, 1/3, and 2/3, we can write the model as:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ReLU(h):
  return h.clip(0.0)

# Three fixed weights and biases for first layer
def h1(x):
  return ReLU(x) # joint at x=0

def h2(x):
  return ReLU(x-1/3) # joint at x=1/3

def h3(x):
  return ReLU(x-2/3) # joint at x=2/3

def model(x, w1, w2, w3, b):
  return b + w1*h1(x) + w2*h2(x) + w3*h3(x)
```

```python
def plot(w1, w2, w3, b):
  x = np.linspace(0, 1, 100)
  y = model(x, w1, w2, w3, b)
  plt.plot(x, y)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.xlim(0, 1)
  plt.ylim(-1, 1)
  plt.yticks(np.arange(-1, 1.2, 0.2))
  plt.show()

# Now search for weights and biases that make the model produces given responses
plot(-0.3, -1.5, 2.4, 0.45)
plot(3.5, -4, -2, -0.8)
plot(0.4, -5.33, 9, 0.67)
```
