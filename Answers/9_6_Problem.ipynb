Plot of the regularization loss term for L0, L1/2, and L1 regularization.

```python
import numpy as np
import matplotlib.pyplot as plt

def loss(x, dim):
  return abs(x[0])**dim + abs(x[1])**dim

# plot the loss function
def plot(x1, x2, dim):
  X1, X2 = np.meshgrid(x1, x2)
  Z = loss([X1, X2], dim)
  fig = plt.figure()
  fig.suptitle(f'Regularization with p={dim}')
  ax = fig.add_subplot(111)
  ax.pcolor(X1, X2, Z, cmap='copper')
  ax.contour(X1, X2, Z, levels=10, colors='gray', linewidths=0.5)
  plt.colorbar(ax.pcolor(X1, X2, Z, cmap='copper'))
  plt.show()

# 2D input space
x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-1, 1, 100)

plot(x1, x2, 0)
plot(x1, x2, 0.5)
plot(x1, x2, 1)
```
