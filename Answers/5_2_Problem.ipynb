Plot of $$L = - \log [1 - \text{sig}[f[\textbf{x}, \bm{\phi}]]]$$ w.r.t. $$\text{sig}[f[\textbf{x}, \bm{\phi}]] \in [0, 1]$$

```python
import numpy as np
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
plt.xlim([0, 1])
ax.set_xlabel('sig[f[x, phi]]')
ax.set_ylabel('L')

x = np.arange(0, 1, 0.01)
y = - np.log(1 - x)

plt.plot(x, y)
```
